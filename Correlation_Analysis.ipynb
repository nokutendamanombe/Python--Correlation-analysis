```python
# First let's import the packages we will use in this project

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

# Now we need to read in the data
df = pd.read_csv(r'C:\Users\nokut\Downloads\archive\movies.csv')
```

```python
# Now  lets look at the data 
df.head()
```

```python
#Now let's check if we have any missing data 
# we will create a for loop to loop through all the columns and check for missing data 

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
#Here we drop any rows with null values 
df = df.dropna(how='any',axis=0)
```

```python
# Here we repeat the check for null values 
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
#Perfect now that we have our null values removed
#Let's check what datatypes we are working with

df.dtypes

```

```python
#changing gross and budget to have rounded integer values 
df['budget'] = df['budget'].astype('int64')
df['gross'] = df ['gross'].astype('int64')

df
```

```python
#order the data by gross revenue

df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Remove Duplicates 

df.drop_duplicates()
```

```python
# Hypothesis 
# Budget has a high correlateion to Gross Revenue
# Production Company has a high correlation to Gross Revenue
```

```python
#Corelation between budget and Gross Revenue Scatter Plot 
sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.ylabel ('Gross Earnings')
plt.xlabel ('Budget for film')

plt.show()
           
```

```python
#Let's start looking at the correlation
#Using the Pearson Correlation Matrix
correlation_matrix = df.corr(numeric_only='TRUE')
correlation_matrix
```

```python

#Let's change the object data types to numerical categories so that we can do a correlation analysis on object data types like company



df_numerised = df.copy()

for col_name in df_numerised.columns:
    if df_numerised[col_name].dtype == 'object':  # Check for object (string) columns
        df_numerised[col_name] = df_numerised[col_name].astype('category')
        df_numerised[col_name] = df_numerised[col_name].cat.codes  # Convert to numerical codes
```

```python
#Let's see the correlation coefficient for everything against Gross Revenue 
correlation_matrix = df_numerised.corr()
print(correlation_matrix['gross'].sort_values(ascending=False))
```

```python
# Correlation with 'gross' column
gross_corr = df_numerised.corr()['gross'].sort_values(ascending=False)

# Bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=gross_corr.index, y=gross_corr.values, palette="viridis")
plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels for readability
plt.title("Correlation with Gross Revenue")
plt.xlabel("Features")
plt.ylabel("Correlation Coefficient")
plt.show()
```
